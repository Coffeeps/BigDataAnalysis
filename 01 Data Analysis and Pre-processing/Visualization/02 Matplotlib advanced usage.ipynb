```python
# plt.plot.box()
fig, axes = plt.subplots(2, 1, figsize=(10,6))
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
color = dict(boxes='DarkGreen', whiskers='DarkOrange', medians='DarkBlue', caps='Gray')

# 
# boxes  
# whiskers  error bar
# medians  
# caps  error bar

df.plot.box(ylim=[0,1.2],
            grid = True,
            color = color,
            ax = axes[0])

df.plot.box(vert=False, # vertTrue
            positions=[1, 4, 5, 6, 8],# position
            ax = axes[1],
            grid = True,
            color = color)
```

```python
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])

plt.figure(figsize=(10,4))

f = df.boxplot(sym = 'o',  # marker
               vert = True,  # 
               whis = 1.5,  # IQR1.5[5,95]95%5%
               patch_artist = True,  # True
               meanline = False,showmeans=True,  # 
               showbox = True,  # 
               showcaps = True,  # 
               showfliers = True,  # 
               notch = False,  # 
               return_type='dict'  # 
              ) 
plt.title('boxplot')

# boxes, 
for box in f['boxes']:
    box.set( color='b', linewidth=1)      # 
    box.set( facecolor='b' ,alpha=0.5)    # 
 
# whiskers, boxerror bar.
for whisker in f['whiskers']:
    whisker.set(color='k', linewidth=0.5,linestyle='-')

# caps, error bar    
for cap in f['caps']:
    cap.set(color='gray', linewidth=2)
    
# medians, ,
for median in f['medians']:
    median.set(color='DarkBlue', linewidth=2)
    
# fliers, 
for flier in f['fliers']:
    flier.set(marker='o', color='y', alpha=0.5)
```
